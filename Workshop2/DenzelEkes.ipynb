```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
#print(soup.prettify())
```

```python
consoles = soup.find(id = "tableID")
one = consoles.find(class_ = "product-info")
print(one.prettify())
#stores = consoles.find_all(class_="sorting_2")
#for store in stores:
 #   print(store.get_text())
```
